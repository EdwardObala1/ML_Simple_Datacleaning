# Filtering Data so as to have and work with only Rwandan Dataset

````python
tax_revenue=pd.read_csv("data-tax-revenue-for-sub-saharan-africa-1980-2010 - Base_All.csv")
rwanda_data = tax_revenue["Country"].str.contains('Rwanda', regex=False)
filtered_data = tax_revenue[rwanda_data]
filtered_data
````

# Analysis for different types of taxes in Rwanda from the year 1980 to year 2010

````python
# list of all the sources of data
Industry = ["Corporate Tax","Direct Taxes", "Individual Tax", "Indirect Taxes", "Trade Taxes", "Non-Resource Taxes" ]
taxsums =[] # total sum of taxes from year 1980 - 2010

def tolist(datas): # this function changes the data from objects to string data types
    listy = []
    for i in datas:
        listy.append(i)
    return listy

def getsum(data): # sums out the taxes on each industry
    sums = 0
    for i in data:
        sums+= float(i)
    return sums

for data_list_object in Industry: #looping through the industry 
    listy = tolist(filtered_data[data_list_object]) # take each column on the data set 
    sums = getsum(listy) #sum the column data 
    taxsums.append(sums) #append the results of column data to a list "taxsums"
plt.bar(Industry, taxsums, align='center', alpha=1)
plt.xlabel('Industry')
plt.ylabel('Tax Scores')
plt.title('Data set showing which industries provide what percentage to the Rwandan Economy ')
````

## Corporate tax data trend for Rwanda 

````python
year = tolist(filtered_data["Year"])
Corporate_tax_Data = tolist(filtered_data["Corporate Tax"])
plt.bar(year, Corporate_tax_Data, align='center', alpha=1, color='c',)
plt.xlabel('year')
plt.ylabel('Taxes collected')
plt.title('Corporate tax data for rwanda since 1980')
````

## Non-Resource Taxes trend for Rwanda 

````python
year = tolist(filtered_data["Year"])
Non_Resource = tolist(filtered_data["Non-Resource Taxes"])
plt.bar(year, Non_Resource, align='center', alpha=1, color='g',)
plt.xlabel('year')
plt.ylabel('Taxes collected')
plt.title('Non-Resource Taxesdata for rwanda since 1980')
````
